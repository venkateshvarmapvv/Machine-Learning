```python
import numpy as np
import pandas as pd
```

```python
data  = pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
out = data.Exited
```

```python
inp = data.drop(["RowNumber", "CustomerId", "Surname","Exited"],1)
```

```python
inp = pd.get_dummies(inp,drop_first = True)
inp.head()
```

```python
from sklearn.preprocessing import RobustScaler
sc = RobustScaler()
sc_data = sc.fit_transform(inp)
sc_data = pd.DataFrame(sc_data)
sc_data.columns = inp.columns
sc_data
```

```python
data.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(inp,out,test_size=0.3,stratify = out,random_state = 48)
```

```python
lm_mod = LogisticRegression()
```

```python
lm_mod.fit(xtrain,ytrain)
```

```python
ypred = lm_mod.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score1 = accuracy_score(ypred,ytest)
confusion = confusion_matrix(ytest,ypred)
accuracy_score1
```

```python
confusion
```

```python
TP = confusion[0,0]
TN = confusion[1,1]
FP  = confusion[1,0]
FN = confusion[0,1]
print(TP,TN,FP,FN)
```

```python
sens = (TP/(TP+FN))*100
spec = (TN/(TN+FP))*100
accu = (TP+TN)/(TP+TN+FP+FN)*100
pre = (TP/(TP+FP))*100
print(sens)
print(spec)
print(accu)
print(pre)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import GridSearchCV
folds = KFold(n_splits = 5,shuffle=True,random_state=100)
logmod = LogisticRegression()
params = {'C':[0.01,0.1,1,10,100],'penalty':['l1','l2']}
grd_mod = GridSearchCV(logmod,params,scoring= 'accuracy',cv = folds)
grd_mod.fit(xtrain,ytrain)
grd_mod.best_params_
```
