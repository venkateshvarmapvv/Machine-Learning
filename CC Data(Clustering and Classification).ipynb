```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('CC_DATA.csv')
```

```python
data.head()
```

```python
data = data.set_index("CUST_ID")
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data_sc = sc.fit_transform(data.dropna())
data_sc = pd.DataFrame(data_sc)
data_sc.columns = data.columns
```

```python
data= data.reset_index()
```

```python
data_sc["CUST_ID"] = data["CUST_ID"]
```

```python
data_sc = data_sc.set_index("CUST_ID")
```

```python
data_sc.head()
```

```python
data_sc.isnull().sum()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score#returns the distance of each cluster from other cluster
```

```python
sil_score = []
for i in range(2,10):
    clus = KMeans(n_clusters=i)
    clus.fit(data_sc)
    clus_pred = clus.labels_
    sil_score.append(silhouette_score(data_sc,clus_pred))
    sil_score
```

```python
clus_mod = KMeans(n_clusters =2)
clus_mod.fit(data_sc)
clus_mod_pred = clus_mod.labels_
```

```python
clus_mod_pred
```

```python
clus_mod_pred = pd.DataFrame(clus_mod_pred)
```

```python
clus_mod_pred["CUST_ID"] = data["CUST_ID"]
```

```python
clus_mod_pred.head()
```

```python
clus_mod_pred = clus_mod_pred.set_index("CUST_ID")
```

```python
data_sc = pd.merge(data_sc,clus_mod_pred, on = "CUST_ID")
```

```python
data_sc.head()
```

```python
out = data_sc[0]
```

```python
inp = data_sc.drop(0,1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(inp ,out, test_size =0.3,random_state = 48)
```

```python
ytest.shape
```

```python
from catboost import CatBoostClassifier
mod_cat = CatBoostClassifier()
mod_cat.fit(xtrain,ytrain)
mod_pred = mod_cat.predict(xtest)
mod_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(ytest,mod_pred))
accuracy_score(ytest,mod_pred)
```
