```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('market_data.csv')
```

```python
out = data.Item_Outlet_Sales
```

```python
data.Item_Weight = data.Item_Weight.fillna(12)
```

```python
data.Outlet_Size = data.Outlet_Size.fillna("Medium")
```

```python
data.Outlet_Establishment_Year = data.Outlet_Establishment_Year.max() - data.Outlet_Establishment_Year
```

```python
inp = data.drop(columns= ["Item_Outlet_Sales","Item_Identifier"], axis =1)
```

```python
inp1 = data.drop(columns= "Item_Outlet_Sales", axis =1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain,ytest = train_test_split(inp1, out, test_size = 0.3, random_state = 48)
```

```python
cat_feat_ind = np.where(inp1.dtypes != np.float)[0]
cat_feat_ind
```

```python
from catboost import CatBoostRegressor
model = CatBoostRegressor()
model.fit(xtrain,ytrain,cat_features = cat_feat_ind)
```

```python
ypredcat = model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(ytest,ypredcat)
mse
```

```python
rsme = mse**0.5
```
